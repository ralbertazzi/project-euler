# Problem 44

### Pentagon numbers

Pentagonal numbers are generated by the formula, Pn=n(3n−1)/2. The first ten pentagonal numbers are:

    1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are pentagonal and D = |Pk − Pj| is minimised; what is the value of D?

### Solution

Brute force solution. It's not guaranteed that the difference is minimised.

```python
import numpy as np

def pentagonal(n):
    return n*(3*n - 1) // 2

pentagonal_numbers = {pentagonal(n) for n in xrange(1, 10000)}
print max(pentagonal_numbers)
```

```python
from tqdm import tqdm

for pj in tqdm(pentagonal_numbers):
    for pk in pentagonal_numbers:
        
        if pj == pk:
            continue
            
        p_sum, p_diff = pj + pk, pk - pj
        
        if p_sum in pentagonal_numbers and p_diff in pentagonal_numbers:
            print pj, pk, abs(pk - pj)
```

A much faster solution, found [online](https://blog.dreamshire.com/project-euler-44-solution/), that builds the pentagon set piece by piece so the search is even faster. The generated number is the sum, and it checks for the middle number and the difference inside the set.

```python
def pe44():
    ps = set()
    i = 1000
    while True:
        i += 1
        s = (3*i*i - i) // 2
        for Pj in ps:
            if s-Pj in ps and s-2*Pj in ps: 
                return s-2*Pj
        ps.add(s)
        
pe44()
```
