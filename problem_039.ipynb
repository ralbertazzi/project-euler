# Problem 39

If p is the perimeter of a right angle triangle with integral length
sides, {a,b,c}, there are exactly three solutions for p = 120.

{20,48,52}, {24,45,51}, {30,40,50}

For which value of p â‰¤ 1000, is the number of solutions maximised?

### Solution

Use an array of counters to store how many times we found a solution for that *p*.
Also compute some square to square root conversions inside the interval of integers required.

```python
import numpy as np

counts = np.zeros(1001, dtype=np.int)
squares = {n**2: n for n in xrange(1000)}
```

If *p* is the perimeter, and *c* the hypotenus, then *a* and *b* will be surely less then *p/2*.
When iterating over *a* and *b*, if we find a squared hypotenuse with square root and the triple of numbers has a perimeter lte 1000, then we have a solution.

```python
for a in xrange(1, 500):
    for b in xrange(a, 500):
        
        c_squared = a**2 + b**2
        if c_squared in squares:
            c = squares[c_squared]
            
            p = a + b + c
            if p <= 1000:
                print (a, b, c)
                counts[p] += 1
```

Print the *p* with the most solutions, and (optional) the number of the solutions

```python
best_p = np.argmax(counts)

print best_p, counts[best_p]
```
