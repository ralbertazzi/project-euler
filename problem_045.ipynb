### Triangular, pentagonal, and hexagonal

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

    Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...
    Pentagonal	 	Pn=n(3n−1)/2	1, 5, 12, 22, 35, ...
    Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...
It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.

### Solution

The proposed solution is very fast and iterates over the hexagonal numbers (because we need less of them -> less comparisons); at each iteration we try to increment the corresponding pentagonal and triangular numbers, untile they are greater or equals to the current hexagonal numbers. A solution is found when all the three numbers are the same.

Notice that no cache (set with all generated numbers) is needed; in fact, we only keep track of the last generated number for each series.

```python
n_t, n_p, n_h = 0, 0, 0
max_t, max_p = 0, 0
count = 0

while True:
    
    n_h += 1
    hexagonal = n_h * (2*n_h - 1)
    
    while hexagonal > max_t:
        
        n_t += 1
        max_t = n_t * (n_t + 1) // 2
        
        
    while hexagonal > max_p:
        
        n_p += 1
        max_p = n_p * (3 * n_p - 1) // 2 
        
    if hexagonal == max_p and hexagonal == max_p:
        
        print hexagonal, n_t, n_p, n_h
        count += 1
        
        if count == 3:
            break
    
```
