### Number letter counts

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

### Solution

First let's write a dictionary of words. I used the words and not the length of the word because I thought that this approach could be less error prone.

```python
digits = {1: 'one', 2: 'two', 3: 'three', 4: 'four', 5: 'five', 6: 'six', 7: 'seven', 8: 'eight', 9: 'nine'}
ten_to_nineteen = {10: 'ten', 11: 'eleven', 12: 'twelve', 13: 'thirteen', 14: 'fourteen', 15: 'fifteen', 
                   16: 'sixteen', 17: 'seventeen', 18: 'eighteen', 19: 'nineteen'}
tens = {10:'ten', 20: 'twenty', 30: 'thirty', 40: 'fourty', 50: 'fifty', 60: 'sixty', 70: 'seventy', 80: 'eighty', 90: 'ninety'}
hundred = 'hundred'
thousand= 'thousand'
```

This first method is accountable to generate the word for numbers between 1 and 99. Notice that a space is generated instead of the character '-', so that we will need to remove only spaces.

```python
def to_words_lt_100(n):
    
    mod_10 = n % 10
    mod_100 = n % 100 - mod_10
    
    if mod_100 == 0:
        return digits[mod_10]
    
    if mod_100 > 0 and mod_10 == 0:
        return tens[mod_100]
    
    if mod_100 == 10:
        return ten_to_nineteen[10 + mod_10]
    else:
        return tens[mod_100] + ' ' + digits[mod_10]
```

This second method is accountable to generate the word for numbers between 100 and 999. The previous method will be called if there's any rest mod 100.

```python
def to_words_gte_100(n):
    
    mod_100 = n % 100
    
    word = ''
    
    h = (n - mod_100) / 100
    word += digits[h] + ' ' + hundred
    if mod_100 > 0:
        word += ' and '
        word += to_words_lt_100(mod_100)
      
    print word
    return word
```

```python
sum_all_numbers = sum(len(to_words_lt_100(n).replace(' ', '')) for n in range(1, 100))
sum_all_numbers += sum(len(to_words_gte_100(n).replace(' ', '')) for n in range(100, 1000))
sum_all_numbers += len('onethousand')
```
