### Powerful digit counts

The 5-digit number, 16807=7^5, is also a fifth power. Similarly, the 9-digit number, 134217728=8^9, is a ninth power.

How many n-digit positive integers exist which are also an nth power?

### Solution

We are looking for numbers of type $x^n, x > 0, n > 0$. If we want that number to have $n$ digits, we can add a simple constraint:

$x^n < 10^n \rightarrow x < 10$

So we know that a *n-digit* number which is also a *n-th power* is generated by a single digit number to the power of *n*.

Also, we need to find when a $x^n, n < 10$ will be too small to generate a n-digit number. This is going to happen when:

$x^n < 10^{n-1}$

We can compute that while finding the powers.

```python
bases = list(range(1, 10))
powers = []
n = 0
```

```python
while len(bases) > 0:
    
    n += 1
    ten_pow_n_minus_1 = 10 ** (n-1)
    
    results = [(b**n, b) for b in bases]
    
    try:
        new_powers, bases = map(list, zip(*[(r, b) for (r, b) in results if r >= ten_pow_n_minus_1]))
    except:
        break
    
    powers += new_powers
    
    print n, bases
    print new_powers
    print
```
